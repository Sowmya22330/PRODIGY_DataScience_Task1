```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample data
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Age': [23, 45, 31, 35, 42, 22, 33, 25, 40, 29]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Bar Chart for Categorical Variable (Gender)
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram for Continuous Variable (Age)
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=5, kde=True, color='blue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# New age distribution data
age_data = {
    'Age': [15, 22, 28, 34, 42, 47, 53, 60, 67, 72],
    'Frequency': [120, 180, 340, 460, 320, 220, 160, 110, 70, 30]
}

# New gender distribution data
gender_data = {
    'Gender': ['Male', 'Female', 'Non-binary'],
    'Frequency': [420, 430, 70]
}

# New civil status distribution data
civil_status_data = {
    'Civil Status': ['Single', 'Married', 'Divorced', 'Widowed', 'Separated'],
    'Frequency': [320, 490, 120, 60, 30]
}

# New nationality distribution data
nationality_data = {
    'Nationality': ['USA', 'UK', 'Canada', 'Australia', 'India', 'Brazil'],
    'Frequency': [260, 210, 140, 110, 310, 420]
}

# New province population data
province_population_data = {
    'Province': ['Province X', 'Province Y', 'Province Z', 'Province W'],
    'Population': [550000, 720000, 430000, 630000]
}

# New marital status distribution data
marital_status_data = {
    'Country': ['USA', 'UK', 'Canada', 'Australia', 'India', 'China'],
    'Single': [160, 110, 85, 75, 210, 260],
    'Married': [210, 160, 130, 105, 260, 320],
    'Divorced': [60, 35, 25, 15, 60, 110],
    'Widowed': [25, 25, 15, 5, 25, 55]
}

# Plotting Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(age_data['Age'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Gender Distribution
gender_df = pd.DataFrame(gender_data)
plt.figure(figsize=(8, 5))
plt.bar(gender_df['Gender'], gender_df['Frequency'], color=['navy', 'lightpink', 'darkgreen'])
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Civil Status Distribution
civil_status_df = pd.DataFrame(civil_status_data)
plt.figure(figsize=(8, 5))
plt.bar(civil_status_df['Civil Status'], civil_status_df['Frequency'], color='tomato')
plt.title('Civil Status Distribution', fontsize=16)
plt.xlabel('Civil Status', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Nationality Distribution
nationality_df = pd.DataFrame(nationality_data)
plt.figure(figsize=(8, 5))
plt.bar(nationality_df['Nationality'], nationality_df['Frequency'], color='teal')
plt.title('Nationality Distribution', fontsize=16)
plt.xlabel('Nationality', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Population Distribution Across Provinces
province_population_df = pd.DataFrame(province_population_data)
plt.figure(figsize=(8, 5))
plt.bar(province_population_df['Province'], province_population_df['Population'], color='slateblue')
plt.title('Population Distribution Across Provinces', fontsize=16)
plt.xlabel('Province', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Marital Status Distribution in Different Countries
marital_status_df = pd.DataFrame(marital_status_data)
marital_status_df.set_index('Country', inplace=True)
marital_status_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'orange', 'grey', 'red'])
plt.title('Marital Status Distribution in Different Countries', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
